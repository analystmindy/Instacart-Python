# [Step 6] Now that you have information about customers, you need to conduct some exploratory analysis of customer demographics to inform the targeted marketing campaigns. First, determine whether there’s a connection between age and family situation by creating a line chart exploring the connections between age and number of dependents. In a markdown cell beneath your line chart, describe what the line chart shows.

```python
# Using the small sample set again to answer the above inquiry about whether there is a connection between age and family
# situation. Create a new df based on these criteria. 

df_4 = small[['age', 'total_dependants']]
```

```python
# Create line chart using the above df

line_age = sns.lineplot (data = df_4, x = 'age', y = 'total_dependants')
```

```python
# Export the line chart

line_age.figure.savefig(os.path.join (path,'04 Analysis', 'Visualizations', 'age_dependants_connection.png'))
```

```python
# There's no clear connection between age and number of dependants. Some portion made sense, and then some didn't. For example,
# it made sense for there to be a dip in number of dependents in the mids-20s age group, then a sudden spike in the 30s as this
# is a normal age group for having children. It dips again in the 40s age group, which presumably are around the time when those
# who had children in theirs 20s would have grown children who may have moved out. It made sense again for there to be a spike
# in the 50s and 60s age group for people who may be taking care of their aging parents. After that, it goes up and down almost
# randomly. Everyone is going to be in a unique situation -- some children move out right at 18, while some stay with their
# parents much longer. Some people lose their parents early on, so they aren't taking care of an "aging parent." Some children
# develop conditions later in life that make them unable to take care of themselves, so they might move back with their parents
# when they're in their 60s. It's all so random everyone's unique situation, which the data supports, so there's no clear cut
# pattern or connection we can base any marketing campaign on. We could do multiple marketing campaigns targeting the different
# circumstances someone may face at different stages of their lives, maybe?
```
