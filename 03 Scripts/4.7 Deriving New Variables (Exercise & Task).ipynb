```python
# The colon (:) before the number indicates the subset should include everything from the begging up to that number.
# The colon (:) after the number indicates the subset should include everything from that number to the end. 

df = ords_prods_merge[:1000000]
```

# [Step 1] Create user-defined function

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

# If-Statements with the loc() Function

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

# If-Statements with For-Loops

```python
ords_prods_merge['order_day_of_week'].value_counts(dropna = False)
```

```python
result = []

for value in ords_prods_merge["order_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

# [Step 2]

```python
result = []

for day in ords_prods_merge['order_day_of_week']:
  if day == 0 or day == 1:
    result.append('Busiest days')
  elif day == 4 or day == 3:
    result.append('Slowest days')
  else:
    result.append('Regular days')
```

```python
ords_prods_merge['busiest_days'] = result
```

# [Step 4]

```python
hours = []

for hour in ords_prods_merge['order_hour_of_day']:
  if hour in [10, 11, 14, 15, 13, 12, 16, 9]:
    hours.append('Most orders')
  elif hour in [23, 6, 0, 1, 5, 2, 4, 3]:
    hours.append('Fewest orders')
  else:
    hours.append('Average orders')
```

```python
# Create the 'busiest_period_of_day' column
ords_prods_merge['busiest_period_of_day'] = hours
```

```python
# Check to see the new columns got added properly
ords_prods_merge.head()
```
